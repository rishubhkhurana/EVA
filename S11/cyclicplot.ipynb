## Cyclic Policy Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
from functools import partial
```

```python
def plot_cycles(n_cycles,period,max_val,min_val,figsize=(16,8),savefig=True):
    out = np.zeros(n_cycles*period)
    def val(max_val,min_val,pos):
        #print(pos,max_val,min_val)
        if pos<0.5:
            return min_val + 2*(max_val-min_val)*pos
        else:
            return max_val + 2*(min_val-max_val)*(pos-0.5)
    fval = partial(val,max_val,min_val)
    #print(fval(0))
    for cyc in range(n_cycles):
        start_index = cyc*period
        for it in range(period):
            pos = it/period
            out[start_index+it] = fval(pos)
    plt.figure(figsize=figsize)    
    plt.plot(range(n_cycles*period),out)
    plt.savefig('cycles.png')
    plt.show()
```

```python
plot_cycles(10,100,10,1)
```
