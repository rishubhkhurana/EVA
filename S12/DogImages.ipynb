## Dog Image Downloader

```python
import urllib
import requests
import os
import json
from skimage.io import imread,imsave
from skimage.transform import resize
import cv2
```

```python
random_url = 'https://dog.ceo/api/breeds/image/random'
```

```python
os.makedirs('dogimages',exist_ok=True)
```

```python
n_images = 200
savepath = './dogimages'
for i in range(n_images):
    img_url = json.loads(requests.get(random_url).content)["message"]
    response = requests.get(img_url)
    file = open(savepath+f'/dog_{i}.jpg','wb')
    file.write(response.content)
    file.close()
```

```python
for i in range(n_images):
    img=imread(savepath+f'/dog_{i}.jpg').astype('uint8')
    image_resized = cv2.resize(img, (224, 224),
                       interpolation = cv2.INTER_AREA)
    imsave(savepath+f'/resized_dog_{i}.jpg',image_resized)
```
